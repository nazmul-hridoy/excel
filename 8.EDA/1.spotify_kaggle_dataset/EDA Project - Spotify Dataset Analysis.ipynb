```python
top_ten_instrumental_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", 
                                                                                           ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
interest_feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
#The target variable of a dataset is the feature of a dataset 
#about which you want to gain a deeper understanding. 
#A supervised machine learning algorithm uses historical data to learn patterns 
#and uncover relationships between other features of your dataset and the target.

interest_feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]
    
    plt.figure(figsize=(6, 4))
    
    
    sns.histplot(pos_data, bins=30, label="Positive", color="green")
    sns.histplot(neg_data, bins=30, label="Negative", color="red")
    
    
    plt.legend(loc="upper right")
    plt.title(f"Positive And Negative Histogram Plot For {feature_col}")
    plt.show()
```

```python
top_ten_energetic_tracks = df[["energy", "song_title", "artist"]].sort_values(by="energy", 
                                                                                           ascending=False)[:10]
top_ten_energetic_tracks
```

```python
plt.figure(figsize=(7,4))
plt.pie(x="energy", data=top_ten_energetic_tracks, autopct='%1.2f%%', labels=top_ten_energetic_tracks.song_title)
plt.show()
```
